```python
import pandas as pd
import numpy as np
import time

from utils import BatchGenerator, batch_balance_sampling, read_img
from callbacks import SlackLogger #custom file for callbacks
from models import unet
```

```python
data = pd.read_pickle("train-data-filelist.pkl")
fms = [read_img(path) for path in data['fissuremask'].values]
```

```python
counts = []
for f in fms:
    _, c = np.unique(f, return_counts=True)
    c = c.tolist()
    if len(c) < 3:
        c.append(0)
    counts.append(c)
```

```python
counts = np.array(counts)
print(counts)
means = np.sum(counts,axis=0)/np.sum(counts)
print(means)
```

```python
weights = 1.0/means
print(weights)
```
